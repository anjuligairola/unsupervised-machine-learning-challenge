```python
# Import Dependencies 
import pandas as pd
import numpy as np
```

```python
# Import the data
myopia_df = pd.read_csv("myopia.csv")
myopia_df.head()
```

```python
#Make a copy of lending_data_d
df=myopia_df.copy()
df
```

```python
#Transform the data
#removing missing or null values
df.dropna(inplace=True)
df
```

```python
#Transform the data
#removing duplicates
df.drop_duplicates(inplace=True)
df
```

```python
#Remove target column to create data for unsupervised model
labels=df["MYOPIC"]
data = df.drop("MYOPIC", axis=1)
data
```

```python
#standardized dataset
#Dependency
from sklearn.preprocessing import StandardScaler
# Scale the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(data)
scaled_features.shape
#till now the data has 618 rows and 14 columns i.e. 14 features
```

```python
 #Dimensionality reduction with PCA
#Dependency
from sklearn.decomposition import PCA
pca = PCA(n_components=0.9)
# Fit the model to the data
data_pca = pca.fit_transform(data)
data_pca.shape
#After employing PCA the features reduced from 14 to 3. So the dimensions were reduced as expected
```

```python
# Further dimensionality reduction with TSNE
#Dependency
from sklearn.manifold import TSNE
# Initialize the t-SNE model with hyperparameters
tsne = TSNE(learning_rate=200)
# Fit the model to the data
data_tsne = tsne.fit_transform(data_pca)
data_tsne.shape
```

```python
# Plot the results
import matplotlib.pyplot as plt
plt.scatter(data_tsne[:,0], data_tsne[:,1])
plt.show()
```

```python
# Results colored by labels
plt.scatter(data_tsne[:,0], data_tsne[:,1], c=labels)
plt.show()
# The output shows that there were no distinct clusters
```

```python
#Cluster Analysis with K-means
#Dependency
from sklearn.cluster import KMeans

#dataframe for pca data
pca_df = pd.DataFrame(data=data_pca)

inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=42)
    km.fit(data_pca)
    inertia.append(km.inertia_)

# Create the dataframe of inertia and k for elbowplot 
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
# Plot the elbow curve to find the best candidate(s) for k
#Dependency
import matplotlib.pyplot as plt
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow curve for myopia data')
plt.show()
#At K=3 was the last inflection point, after that the inertia values are not changing as drastically
```

```python
def get_clusters(k, data):
    # Initialize the K-Means model
    model = KMeans(n_clusters=k, random_state=42)

    # Train the model
    model.fit(data)

    # Predict clusters
    predictions = model.predict(data)

    # Create return DataFrame with predicted clusters
    data["class"] = model.labels_

    return data
```

```python
#Analyzing Clusters with the Best Value of k
clusters = get_clusters(3,pca_df)
clusters
```

```python
clusters.rename(columns={0:"PCA1",1:"PCA2",2:"PCA3"},inplace=True)
```
